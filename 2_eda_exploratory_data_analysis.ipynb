```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("preprocessed_data.csv")
data.isnull().sum()
```

```python
#now we will be finding some pattern between our required features and the result
data.head()
```

```python
#lets check about how many really got accepted and how many got rejected
final_result = data["loan_status"].value_counts()

final_result.value_counts
```

```python
plt.pie(

    final_result.values,
    labels = final_result.index,
    autopct="%1.1f%%"

)
plt.title("Accepted vs Rejected")
plt.show()

#conclusion is more people are getting approved overall
```

```python
#now checking education level of all those people who got accepted to understand if it a important criteria or not
edu = data[data["loan_status"]=="Approved"].education
edu.value_counts()
```

```python


plt.pie(

    edu.value_counts().values,
    labels = edu.value_counts().index,
    autopct="%1.1f%%"

)
plt.title("education values for approved customers")
plt.show()
```

```python
edu = data[data["loan_status"]=="Rejected"].education
edu.value_counts()

plt.pie(

    edu.value_counts().values,
    labels = edu.value_counts().index,
    autopct="%1.1f%%"

)
plt.title("education values for rejected customers")
plt.show()

# so we can conclude education is not such a useful data
```
